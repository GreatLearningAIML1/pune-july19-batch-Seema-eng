### Plot the attrition column (Since it is a categorical variable, chose the plot accordingly)

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
attr_yes = Data[Data.Attrition=='Yes']
sb.distplot(attr_yes.Age)
plt.ylabel('Attrition = Yes')
plt.show()
```

```python
#Attrition according to Marital status
sb.countplot(x='Attrition', hue='MaritalStatus', data=Data)
plt.show()
```

```python
#Attrition across Job Roles
attr_yes = Data[Data.Attrition=='Yes'] #subset with Attrition = Yes
attr_no = Data[Data.Attrition=='No'] ##subset with Attrition = No
fig, ax6 = plt.subplots(1,2, figsize=(24,6))
sb.countplot(x='Department', hue='Gender', data = attr_yes, ax = ax6[1])
sb.countplot(y='JobRole', hue='Gender', data = attr_yes, ax = ax6[0])
plt.show()
```

```python
Data['PromoYrLvl'] = pd.cut(Data['YearsSinceLastPromotion'], 5, labels=['0to3', '3to6', '6to9', '9to12', '12+'])

sb.countplot(x='PromoYrLvl',hue='Attrition', data=Data)

```

```python
Data['CmpYrInterval'] = pd.cut(Data['YearsAtCompany'], 5, labels=['<9', '<15', '<24', '<32', '<33+'])
sb.countplot(x='CmpYrInterval', hue='Attrition', data =Data)
```

```python
fig, ax4 = plt.subplots(2,2, figsize=(10,10))
sb.boxplot(x='Attrition', y='DistanceFromHome', data=Data, ax = ax4[0,0])
sb.boxplot(x='Attrition', y='PercentSalaryHike', data=Data, ax = ax4[0,1])
sb.boxplot(x='Attrition', y='NumCompaniesWorked', data=Data, ax = ax4[1,0])
sb.boxplot(x='Attrition', y='MonthlyRate', data=Data, ax = ax4[1,1])

plt.show()
```
