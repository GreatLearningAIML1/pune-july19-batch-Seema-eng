## Question 2

Given,

Null Hypothesis : There is no significant difference between data sets

Alternate Hypothesis : There is a significant difference

*Do two-sample testing and check whether to reject Null Hypothesis or not.*

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html

```python
alpha =0.05
Ttest,pvalue=stats.ttest_ind(GPU1,GPU2) #standard independent 2 sample test that assumes equal population variances
#stats.ttest_ind(GPU1,GPU2, equal_var = False) # t-test, which does not assume equal population variance
print(pvalue)
if pvalue<alpha:
    print("Reject Null")
else:
    print("Fail to Reject Null")
```

As the p value is smaller than the threshold, we reject the null hypothesis. This means both GPUs are not effectively the same speed.

The one-way ANOVA tests the null hypothesis that two or more groups have the same population mean. The test is applied to samples from two or more groups, possibly with differing sizes.

use stats.f_oneway() module to perform one-way ANOVA test

*In one or two sentences explain about **TypeI** and **TypeII** errors.*

#### Answer:

No hypothesis test is 100% certain. Because the test is based on probabilities, there is always a chance of making an incorrect conclusion. When you do a hypothesis test, two types of errors are possible: type I and type II.
Type I error
When the null hypothesis is true and reject it, it makes a type I error. 
Type II error
When the null hypothesis is false and fail to reject it, it makes a type II error.

## Question 8

Let's create a small dataset for dice rolls of four players

```python
import numpy as np

d1 = [5, 8, 3, 8]
d2 = [9, 6, 8, 5]
d3 = [8, 12, 7, 2]
d4 = [4, 16, 7, 3]
d5 = [3, 9, 6, 5]
d6 = [7, 2, 5, 7]

dice = np.array([d1, d2, d3, d4, d5, d6])
```

run the test using SciPy Stats library

Depending on the test, we are generally looking for a threshold at either 0.05 or 0.01. Our test is significant (i.e. we reject the null hypothesis) if we get a p-value below our threshold.

For our purposes, weâ€™ll use 0.01 as the threshold.

use stats.chi2_contingency() module 

This function computes the chi-square statistic and p-value for the hypothesis test of independence of the observed frequencies in the contingency table

Print the following:

- chi2 stat
- p-value
- degree of freedom
- contingency



```python
from scipy import stats

stats.chi2_contingency(dice)
```

```python
chi2_stat, p_val, dof, ex = stats.chi2_contingency(dice)
print("   Chi2 Stat    ")
print(chi2_stat)
print("\n")
print("   Degrees of Freedom    ")
print(dof)
print("  P-Value   ")
print(p_val)
print("\n")
print("   Contingency Table    ")
print(ex)
```

Our test is not significant (i.e. we fail to reject the null hypothesis) as we get a p-value equals to 0.07 which isabove our threshold.
Thus we have not met the threshold for statistical significance.

## Question 10

A Paired sample t-test compares means from the same group at different times.

The basic two sample t-test is designed for testing differences between independent groups. 
In some cases, you might be interested in testing differences between samples of the same group at different points in time. 
We can conduct a paired t-test using the scipy function stats.ttest_rel(). 

```python
before= stats.norm.rvs(scale=30, loc=100, size=500) ## Creates a normal distribution with a mean value of 100 and std of 30
after = before + stats.norm.rvs(scale=5, loc=-1.25, size=500)
```

Test whether a weight-loss drug works by checking the weights of the same group patients before and after treatment using above data.

```python
import  scipy.stats  as stats  
stats.ttest_rel(before,after)
```

p value 7.9416e-12 < 0.05 (alpha) so we reject null , means their is significance difference  in the weights of the same group patients before and after treatment.
