# Function which returns all the team with their goal differences 
#team which has max and min goal difference.

```python
def Goal_diff_count():
    diff=df['GoalsFor'].astype(int)-df['GoalsAgainst'].astype(int)
    
    return(diff)

diff = Goal_diff_count()
max_val = 0
min_val = 0
max_team = ""
min_team = ""
print("goal difference of each team :\n")
for i in range(0,len(diff)):
    print(df['Team'][i]+"\t\t\t"+str(diff[i]))
    
    if max_val <int(diff[i]):
        max_val = int(diff[i])
        max_team = df['Team'][i]
    if min_val > int(diff[i]):
        min_val = int(diff[i])
        min_team = df['Team'][i]
        
print ("Max: ",max_team, max_val)
print("Min: ", min_team, min_val)
```
